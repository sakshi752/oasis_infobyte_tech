# preprocessing of data

```python
iris_df
```

```python
iris_df[iris_df['Species']=='Iris-versicolor']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
iris_df['Species']=le.fit_transform(iris_df['Species'])
# Iris-setosa - 0 , Iris-versicolor - 1 , Iris-virginica-2
```

- soo each of 50 data for each species

```python
iris_df['Species'].value_counts()
```

- dividing the dataset into dependent and independent variables

```python
X=iris_df.drop('Species',axis=1)  # Features (input columns)
y=iris_df['Species']  #target col
```

- splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
# dividing the dataset in training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
# determining the shapes of training and testing sets
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Classification Algorithm (supervised learning) 

```python
from IPython.display import Image
Image(filename='Supervised_ML.png')
```

1. SVM

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

2. Random Forest

```python
# Create an instance of th e RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
# evaluate the performance of the model
from sklearn.metrics import accuracy_score
accuracy1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy1)
```

3. KNN
- The object is assigned to the class which is most common among its k (an integer specified by human) nearest neighbour.

```python
from sklearn.neighbors import KNeighborsClassifier
# create an instance of KNN classifier
knn = KNeighborsClassifier(n_neighbors = 7)
# train the model using training data
knn.fit(X_train,y_train)
# test the model on testing data
y_pred=knn.predict(X_test)

# evaluate the performance of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

- new input

```python
def iris_classify(input):
    if input==0:
        print("It is a Iris-setosa")
    elif input==1:
         print("It is a Iris-versicolor")
    else:
         print("It is a Iris-virginica")
```
