# Feature Selection

```python
#from above we get to know features that are highly correlated to car price in +ve way
final_df=df[['price','enginesize', 'curbweight', 'horsepower', 'carwidth', 'highwaympg', 'citympg', 'carlength', 'wheelbase', 'boreratio']]
final_df.head()
```

```python
print("size= ",final_df.size)
print("dimensions= ",final_df.shape)
print(final_df.dtypes)
```

```python
# div dataset to two different df one for independent features and other for dependent features
X=final_df.drop('price',axis=1)
y=final_df['price']
```

```python
from sklearn.model_selection import train_test_split
# dividing the dataset in training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
# determining the shapes of training and testing sets
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```
